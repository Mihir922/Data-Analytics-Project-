```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("/dailyActivity_merged.csv")
```

```python
cols=['Id','ActivityDate','TotalSteps','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes','Calories']
```

```python
df=data[cols]
df
```

```python
df.rename(columns={'ActivityDate':'data'},inplace=True)
```

```python
df
```

```python
df['TotalMinutes'] = df.VeryActiveMinutes	+ df.FairlyActiveMinutes + df.LightlyActiveMinutes + df.SedentaryMinutes
```

```python
df
```

```python
df['TotalHour'] = round(df['TotalMinutes']/60)
df
```

```python
df.info()
```

```python
df['DayofWeek'] = pd.to_datetime(df['data']).dt.day_name()
```

```python
minutes = [df['VeryActiveMinutes'].sum(),
           df['FairlyActiveMinutes'].sum(),
           df['LightlyActiveMinutes'].sum(),
           df['SedentaryMinutes'].sum()]

labels=['veryActiveMinutes','fairlyActiveMinutes','lightlyActiveMinutes','sedentaryMinutes']

plt.pie(minutes, labels=labels, autopct='%1.1f%%', explode=[0,0,0,0.20])
plt.show()
```

```python
# Group data by day of the week and calculate average steps and calories
day_of_week_analysis = df.groupby('DayofWeek')[['TotalSteps', 'Calories']].mean()

# Visualize the results
day_of_week_analysis.plot(kind='bar', figsize=(10, 6))
plt.title('Average Steps and Calories by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the correlation between active minutes and calories
correlation = df[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'Calories']].corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation, annot=True, cmap='hot')
plt.title('Correlation between Active Minutes and Calories')
plt.show()
```

```python
#scatter plot of total steps vs. active minutes
plt.figure(figsize=(10, 6))
plt.scatter(df['TotalSteps'], df['VeryActiveMinutes'] + df['FairlyActiveMinutes'] + df['LightlyActiveMinutes'], color='purple')
plt.title('Relationship between Total Steps and Active Minutes')
plt.xlabel('Total Steps')
plt.ylabel('Active Minutes')
plt.show()
```

```python
# Calculate the interquartile range (IQR) for total steps
Q1 = df['TotalSteps'].quantile(0.25)
Q3 = df['TotalSteps'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Identify outliers
outliers = df[(df['TotalSteps'] < lower_bound) | (df['TotalSteps'] > upper_bound)]

# Print the outliers
print(outliers)
```
