# AIRBNB DATA ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("compressed_data.csv")
df.head(3)
```

# Check the column names in the Dataset

```python
df.columns
```

# Check the Missing Values 

```python
print(df.isnull().sum())
```

# Handling Missing Values 

this code ensures that  the "the last Review" column is properly formatted as datetime,missing values in key columns are apporopriately handled and incompleted records are removed, preparing the Dataset for further analysis or Visualization.

```python
df['last review'] = pd.to_datetime(df['last review'],errors='coerce')

df.fillna({'reviews per month':0,'last review': df['last review'].min()},inplace=True)
          
df.dropna(subset=['NAME','host name'],inplace=True)
```

```python
print(df.isnull().sum())
```

# Correct Data Type 

Ensure the all columns have the correct data types. 

```python
df['price'] = df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

# Remove Duplicates 

```python
df.drop_duplicates(inplace=True)
```

# Confirme Data Cleaning 

```python
print(df.info())
```

```python
df = df.drop(columns=["license","house_rules"],errors="ignore")
```

 # Visulization 

# Destribution of Prices

Plot the destribution of listing prices 

```python
plt.figure(figsize=(10,6))
sns.histplot(df["price"],bins=50,kde=True,color="red")
plt.title("Distribution of Listing Prices")
plt.xlabel("Prices ($)")
plt.ylabel("Frequency")
plt.show()
```

# Room Type Analysis

# Analyze the Distribution of different room types 

```python
plt.figure(figsize=(8,5))
sns.countplot(x="room type",data=df,color="red")
plt.title("Room Type Distribution")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.show()
```

# Neighborhood Analysis 

# Examine How Listings are distributed across different neighborhoods

```python
plt.figure(figsize=(12,8))
sns.countplot(y="neighbourhood group",data=df,color="red",order=df["neighbourhood group"].value_counts().index)
plt.title("Number of listings by neighbourhood group")
plt.xlabel("Count")
plt.ylabel("Neighbourhood group")
plt.show()
```

# Price Vs Room Type

# Visulization for the Relationship between Price and Type 

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="room type",y="price",hue="room type",data=df,palette="Set3")
plt.title("Price VS Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price ($)")
plt.legend(title="Room Type")
plt.show()
```

# Revies Over Time 

```python
df["last review"] = pd.to_datetime(df["last review"])
review_over_time =df.groupby(df["last review"].dt.to_period("M")).size()

plt.figure(figsize=(12,6))
review_over_time.plot(kind="line",color="red")
plt.title("Number Of Review Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Review")
plt.show()
```
